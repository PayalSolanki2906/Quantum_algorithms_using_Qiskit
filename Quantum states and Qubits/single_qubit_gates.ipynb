```python
from qiskit import QuantumCircuit, assemble, Aer
from math import pi, sqrt, exp
from qiskit.visualization import plot_bloch_multivector, plot_histogram
sim = Aer.get_backend('aer_simulator')
```

# The Pauli-X

The Pauli-X gate acts on a single-qubit state. This is the quantum gate equivalent to the NOT gate. It is sometimes also known as a bit flip operator. It maps $|0\rangle$ to $|1\rangle$ and $|1\rangle$ to $|0\rangle$.

$$X=\left(
\begin{matrix}
0 & 1\\
1 &0
\end{matrix} \right).$$

## Truth table

| Input  | Output | 
| --- | --- |
| $\vert0\rangle$ | $\vert1\rangle$ |
| $\vert1\rangle$ | $\vert0\rangle$ |


## Let's do an X-gate on a |0> qubit


```python
qc = QuantumCircuit(1)
qc.x(0)
qc.draw()
```

```python
qc.save_statevector()
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(state)
```

# The Pauli-Y

The Pauli-Y gate acts on a single-qubit state. It maps $|0\rangle$ to $\dot{\iota}|1\rangle$ and $|1\rangle$ to $-\dot{\iota}|0\rangle$. It is clear that, the Pauli-Y gate is similar to the Pauli-X gate but with an additional phase of $\pm \dot{\iota}$

$$Y=\left(
\begin{matrix}
0 & -\dot{\iota}\\
\dot{\iota} &0
\end{matrix} \right).$$

## Truth table

| Input  | Output | 
| --- | --- |
| $\vert0\rangle$ | $\dot{\iota}\vert 1\rangle$ |
| $\vert1\rangle$ | $-\dot{\iota}\vert 0\rangle$ |


```python
qc = QuantumCircuit(1)
qc.y(0)
qc.draw()
qc.save_statevector()
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(state)
```

$$Z=\left(
\begin{matrix}
1 & 0\\
0 &-1
\end{matrix} \right).$$

## Truth table

| Input  | Output | 
| --- | --- |
| $\vert 0\rangle$ | $\vert 0\rangle$ |
| $\vert 1\rangle$ | $-\vert 1\rangle$ |


```python
qc = QuantumCircuit(1)
qc.z(0)
qc.draw()
qc.save_statevector()
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(state)
```

# The Hadamard gate-H

The Hadamard gate acts on a single-qubit state. It maps $|0\rangle$ to $\frac{|0\rangle+|1\rangle}{ \sqrt{2}}$ and  $|1\rangle$ to $\frac{|0\rangle-|1\rangle}{ \sqrt{2}}$. The Hadamard gate is often used to interconvert between the Pauli-X and the Pauli-Z eigenstates. It is to be noted that the columns of the Hadamard matrix are the eigenvectors of Pauli-X when represented in the computational basis. 
$$H=\frac{1}{\sqrt{2}}\left(
\begin{matrix}
1 & 1\\
1 &-1
\end{matrix} \right).$$

## Truth table

| Input  | Output | 
| :------------- | -------------: |
| $\vert 0\rangle$ | $\frac{\vert 0\rangle+\vert 1\rangle}{ \sqrt{2}}$ |
| $\vert 1\rangle$ | $\frac{\vert 0\rangle-\vert 1\rangle}{ \sqrt{2}}$ |


```python
qc = QuantumCircuit(1)
qc.h(0)
qc.draw()
qc.save_statevector()
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(state)
```

# The P gate

The P-gate (phase gate) is parametrised, that is, it needs a number ($\phi$) to tell it exactly what to do. The P-gate performs a rotation of $\phi$ around the Z-axis direction. 
$$P=\left(
\begin{matrix}
1 & 0\\
0 &e^{\dot{\iota}\phi}
\end{matrix} \right).$$

## Truth table

| Input  | Output | 
| --- | --- |
| $\vert 0\rangle$ | $\vert 0\rangle$ |
| $\vert 1\rangle$ | $e^{\dot{\iota}\phi} \vert 1\rangle$ |

```python
qc = QuantumCircuit(1)
qc.p(pi/4, 0)
qc.draw()
```

# The S gate

The S-gate is a P gate with $\phi=\pi/2$

$$S=\left(
\begin{matrix}
1 & 0\\
0 & \dot{\iota}
\end{matrix} \right).$$

$$S^\dagger=\left(
\begin{matrix}
1 & 0\\
0 & -\dot{\iota}
\end{matrix} \right).$$

## Truth table for S gate

| Input  | Output | 
| --- | --- |
| $\vert 0\rangle$ | $\vert 0\rangle$ |
| $\vert 1\rangle$ | $\dot{\iota} \vert 1\rangle$ |

```python
qc = QuantumCircuit(1)
qc.s(0)   # Apply S-gate to qubit 0
qc.sdg(0) # Apply Sdg-gate to qubit 0
qc.draw()
```

# The T gate

The T-gate is a P gate with $\phi=\pi/4$

$$T=\left(
\begin{matrix}
1 & 0\\
0 & e^{\dot{\iota}\pi/4}
\end{matrix} \right).$$

$$T^\dagger=\left(
\begin{matrix}
1 & 0\\
0 & e^{-\dot{\iota}\pi/4}
\end{matrix} \right).$$

## Truth table for T gate
 Input     | Output.
 :------------- | --------------: 
 $\vert 0\rangle$ | $\vert 0\rangle$ 
 $\vert 1\rangle$ | $e^{\dot{\iota}\frac{\pi}{4}} \vert 1\rangle$ 

```python
qc = QuantumCircuit(1)
qc.t(0)   # Apply T-gate to qubit 0
qc.tdg(0) # Apply Tdg-gate to qubit 0
qc.draw()
```

# The U-gate

The U-gate is the most generalized version of all single qubit quantum gates.

$$U(\theta,\phi,\lambda)=\left(
\begin{matrix}
\cos\left(\frac{\theta}{2}\right) & e^{-\dot{\iota}\lambda}\sin\left(\frac{\theta}{2}\right)\\
e^{\dot{\iota}\phi}\sin\left(\frac{\theta}{2}\right) & e^{\dot{\iota}(\phi+\lambda)}\cos\left(\frac{\theta}{2}\right)
\end{matrix} \right).$$

Every gate can be decribed using $\theta$, $\phi$ and $\lambda$, for example,
$$U\left(\frac{\pi}{2},0,\pi\right)=\frac{1}{\sqrt{2}}\left(
\begin{matrix}
1 & 1\\
1 &-1
\end{matrix} \right)=H.$$




```python
qc = QuantumCircuit(1)
qc.u(pi/2, 0, pi, 0)
qc.draw()
```

```python
qc.save_statevector()
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(state)
```

# Measuring in Different Bases

The probability of measuring a quantum state $|\psi\rangle$ in a particular state $|\phi\rangle$ is given by the squared modulus of the inner product between the two states, which is written as:

$$P(|\phi\rangle) = |\langle \phi | \psi \rangle|^2$$

This is known as the Born rule.
